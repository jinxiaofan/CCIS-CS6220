```python
# Problem 5
#  user_same_rating that given a user_id and movie_id,
#  returns a list of user_ids 
#  for other users that rated the movie identified by 
#  the provided movie_id with the same score. 
def user_same_rating(uid, mid):
    rating = ratings[(ratings['user_id'] == uid) & (ratings['movie_id'] == mid)]['rating'].values[0]
    users = list(ratings[(ratings['movie_id'] == mid) & (ratings['rating']==rating)]['user_id'])
    users.remove(uid)
    return users
```
