```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
FileUrl = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = pd.read_csv(FileUrl, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'],header=None)
```

```python
# 1.1 Print the first 5 elements of your DataFrame using the command head(). 
pd.DataFrame.head(data,n=5)
```

```python
# Q:How many features are there and what are their types (e.g., numeric, nominal)?
# A:There are four features and one class. Features are numeric and name is nominal(object)
data.dtypes
```

```python
# 1.2 The 8 is variance and the 9 is range
attribute = data.describe()
variance = data.var()
attribute.loc[8] = variance
range = data[data.columns[0:4]].max()-data[data.columns[0:4]].dropna().min()
attribute.loc[9] = range
attribute
```

```python
#variance
data.var()
```

```python
#range
data[data.columns[0:4]].max()-data[data.columns[0:4]].dropna().min()
```

```python
# 1.2 histogram separeted by each column
data.hist()
plt.show()
```

```python
# boxplot separeted by each column
data.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
# 1.2 boxplot together
box = data.boxplot(grid = False)
```
